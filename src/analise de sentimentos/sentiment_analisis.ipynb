```python
# salvando os dataframe em variaveis
import pandas as pd
elon_data = pd.read_csv('data.csv')
data= pd.read_csv('tweets.csv')
```

```python
# gerando a coluna engajamento
elon_data['engagement'] = elon_data['like'] + elon_data['comment'] + elon_data['rt']
data['engagement'] = data['like'] + data['comment'] + data['rt']
```

```python
# algoritimo para a normalização dos textos
import re  
def remove_special_character(tweet):
    # remove the old style retweet text "RT"
    tweet = re.sub(r'^RT[\s]+', '', tweet)

    # remove hyperlinks
    tweet = re.sub(r'https?:\/\/.*[\r\n]*', '', tweet)

    # remove hashtags. only removing the hash # sign from the word
    tweet = re.sub(r'#', '', tweet)

    # remove single numeric terms in the tweet. 
    tweet = re.sub(r'[0-9]', '', tweet)
    
    return tweet
```

```python
# normalizando os textos
elon_data['tweet'] = elon_data['tweet'].apply(lambda tweet: remove_special_character(tweet))
elon_data['tweet'] = elon_data['tweet'].str.lower()
data['tweet'] = data['tweet'].apply(lambda tweet: remove_special_character(tweet))
data['tweet'] = data['tweet'].str.lower()
```

```python
# gerando a analize do afinn
from afinn import Afinn
afinn = Afinn()
data['afinn'] = data['tweet'].apply(lambda tweet: afinn.score(tweet))/20
elon_data['afinn'] = elon_data['tweet'].apply(lambda tweet: afinn.score(tweet))/20
```

```python
# algoritimo para a nota do vader
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
vader = SentimentIntensityAnalyzer()
def getVaderScore(tweet):    
    vader_score = vader.polarity_scores(tweet)
    score = vader_score['compound']
    return score*5
```

```python
# dando a nota do vader para os textos
# data['vander_sentiment'] = data['vander_score'].apply(lambda score: getVanderSentiment(score))
data['vader'] = data['tweet'].apply(lambda tweet: getVaderScore(tweet))
elon_data['vader'] = elon_data['tweet'].apply(lambda tweet: getVaderScore(tweet))
```
